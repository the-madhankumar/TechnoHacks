```python

from keras.models import Sequential
from keras.layers import Dense,Flatten,Conv2D,MaxPool2D,Dropout 

from keras.utils import to_categorical
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
from tensorflow.keras.datasets import cifar10
```

```python
(x_train,y_train),(x_test,y_test) = cifar10.load_data()
```

```python
x_train.shape
```

```python
plt.imshow(x_train[0])
```

```python
x_train.max()
```

```python
x_train[0]
```

```python
classification=['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']
```

```python
print('image class is:',classification[y_train[10][0]])
```

```python
ytrain_one_hot=to_categorical(y_train)
ytest_one_hot=to_categorical(y_test)
```

```python
ytrain_one_hot[10]
```

```python
print('one hot label:',ytrain_one_hot[10])
```

```python
xtrain=x_train/255
xtest=x_test/255
```

```python
xtrain[10]
```

```python

model=Sequential()
model.add(Conv2D(32,(5,5),activation='relu',input_shape=(32,32,3)))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Conv2D(32,(5,5),activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Flatten())

model.add(Dense(1000,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(500,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(250,activation='relu'))

model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss="categorical_crossentropy",optimizer = "adam",metrics=["accuracy"])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss',patience=3)
```

```python
model.fit(x_train,ytrain_one_hot,epochs=15,validation_data=(x_test,ytest_one_hot),callbacks=[early_stop])
```

```python
import pandas as pd
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
losses[['accuracy','val_accuracy']].plot()
```

```python
losses[['loss','val_loss']].plot()
```

```python
model.metrics_names
```

```python
print(model.metrics_names)
print(model.evaluate(x_test,ytest_one_hot,verbose=0))
```
